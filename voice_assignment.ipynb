```python
import whisper
import torchaudio

def load_audio(file_path):
    # Load and preprocess the audio file
    waveform, sample_rate = torchaudio.load(file_path)
    waveform = torchaudio.transforms.Resample(orig_freq=sample_rate, new_freq=16000)(waveform)

    # Convert to mono if necessary
    if waveform.shape[0] > 1:
        waveform = torch.mean(waveform, dim=0, keepdim=True)

    return waveform, 16000
```

```python
def apply_vad(waveform, sample_rate, vad_threshold=0.5):
    # Apply Voice Activity Detection (VAD)
    vad = torchaudio.transforms.Vad(sample_rate=sample_rate)
    return vad(waveform)

def speech_to_text(file_path):
    model = whisper.load_model("base")
    waveform, sample_rate = load_audio(file_path)
    waveform = apply_vad(waveform, sample_rate)

    # Convert speech to text
    transcription = model.transcribe(file_path)
    return transcription["text"]
```

```python
# Example usage
audio_file = r"/content/harvard.wav"
text = speech_to_text(audio_file)
print("Transcribed Text:", text)
```
